```python
import pandas as pd

# Path to your CSV file
file_path = '/content/2022-08-03-ss.cleaned.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to check the content
df.head()
```

```python
# Remove rows where has_nonstd_aa is True
filtered_df = df[df['has_nonstd_aa'] == False]
# Drop the columns 'pdb_id', 'sst8', and 'chain_code'
final_df = filtered_df.drop(columns=['pdb_id', 'sst8', 'chain_code'])
final_df = final_df[(final_df['len'] >= 10) & (final_df['len'] <= 80)]
final_df = final_df.drop(columns=['has_nonstd_aa'])
# Display the first few rows of the final DataFrame
final_df.head()
```

```python
# Check the size of the original DataFrame
num_rows = final_df.shape[0]
print(f"Total rows in final_df: {num_rows}")

# Set the sample sizes
train_sample_size = 1600
test_sample_size = 4000

# Ensure the sample sizes do not exceed the available data
if num_rows < train_sample_size + test_sample_size:
    raise ValueError("Not enough data to sample the specified sizes.")

# Step 1: Randomly sample 16000 rows for training data
train_data = final_df.sample(n=train_sample_size, random_state=42)

# Step 2: Remove the selected training data from the original DataFrame
remaining_data = final_df.drop(train_data.index)

# Step 3: Randomly sample 4000 rows for test/validation data from the remaining data
test_data = remaining_data.sample(n=test_sample_size, random_state=42)
train_data = train_data.drop(columns=['len'])
test_data = test_data.drop(columns=['len'])

# Step 4: Save the training and test/validation data to separate CSV files
train_data.to_csv('/content/training_data_part4.csv', index=False)
test_data.to_csv('/content/test_data_part4.csv', index=False)

# Display a message indicating completion
print("Training and test data saved successfully.")
```
